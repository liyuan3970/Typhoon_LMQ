```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib
print(matplotlib.matplotlib_fname())
path ='../src/data.xlsx'

data = pd.read_excel(path)

#print(data.iloc[:,7])
x =  data.iloc[:,0]
#x = np.arange(1, 54, 1)
# print(x)
y = data.iloc[:,1]
y1 =data.iloc[:,2]
y2 = data.iloc[:,3]
y3 =data.iloc[:,4]
y4 =data.iloc[:,5]
y5 =data.iloc[:,6]
y6 =data.iloc[:,7]
y7 =data.iloc[:,8]

font1 = {'family' : 'Times New Roman',
'weight' : 'normal',
'size'   : 18,
}
plt.rcParams['font.sans-serif'] = ['SimHei']  # 
plt.rcParams['axes.unicode_minus'] = False  # 
labels = ["Vorticity Average", "Vorticity Max"]
# colors = ["#8da0cb", "#fc8d62", "#66c2a5"]
x1 = [0,6,15,22,40,53]
xticks1 =['23:04','23:33','00:29','01:08','02:49','04:02']
plt.figure(figsize=(25,16))
plt.plot(x, y,label='Vorticity Average',linewidth =5)
plt.plot(x, y1,label='Vorticity Max',linewidth =5)
y_values = np.zeros([54,1])
plt.plot(x,y_values,c='black')
plt.grid()
plt.xticks(x1,xticks1,size='large')
plt.xlabel('time')
plt.ylim(-2,4)
plt.legend(loc="upper right",fontsize=30)

plt.show()
```
