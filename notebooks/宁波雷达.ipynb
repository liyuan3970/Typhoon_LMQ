```python
import cinrad
file = '/home/liyuan3970/Data/radar_typhoon_liqima/ningbo_radar/10/'
f = cinrad.io.CinradReader(file+'Z_RADR_I_Z9574_20190810000200_O_DOR_SA_CAP.bin.bz2')
tilt_number = 0
data_radius = 230
data_dtype = 'REF' # stands for reflectivity
r = f.get_data(tilt_number, data_radius, data_dtype)
print(r)


%matplotlib inline
rl = list(f.iter_tilt(230, 'REF'))
#rl = list(f.iter_tilt(230, 'VEL'))
cr = cinrad.easycalc.quick_cr(rl)
fig = cinrad.visualize.PPI(cr, dpi=50)
```
